Importing Libraries

```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import dateutil
```

Using request library ,fretch data from the link

```python
result=requests.get("https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_population")
```

```python
result.status_code
```

```python
assert result.status_code==200
```

```python
print(result.content)
```

Preparing source code of website

```python
sorce=result.content
document=BeautifulSoup(sorce,'lxml')
```

Find 'table' tag in prepared document

```python
table=document.find('table')
```

```python
table.find('th').get_text()==""
```

```python
assert table.find('th').get_text()==""
```

```python
type(table)
```

```python
df=pd.read_html(str(table))
df
```
